```python
# Import dependencies
import pandas as pd
```

```python
# Read the file
file_to_load = ("vine_table.csv")
vine_df = pd.read_csv(file_to_load)

vine_df
```

```python
# Step 1
# Create total_votes dataframe where total_votes count is equal to or 
# greater than 20
total_votes_df = vine_df[vine_df.total_votes >= 20]
total_votes_df
```

```python
# Step 2
# Create a new df filtering total_votes_df where the number of helpful_votes
# divided by total_votes is equal to or greater than 50%
helpful_votes_df = total_votes_df[((total_votes_df["helpful_votes"]/total_votes_df["total_votes"])*100 ) >= 50]
helpful_votes_df
```

```python
# Step 3
# Create a new df filtering df in step 2 that retreives all rows where
# a review was written as part of the Vine program (paid), vine == 'Y'
vine_program_paid = helpful_votes_df[helpful_votes_df.vine == "Y"]
vine_program_paid
```

```python
len(vine_program_paid)
```

```python
# Step 4
# Repeat Step 3, but this time retrieve all the rows where the review was not part 
# of the Vine program (unpaid), vine == 'N'.
vine_program_unpaid = helpful_votes_df[helpful_votes_df.vine == "N"]
vine_program_unpaid
```

```python
# Step 5
# Determine the total number of reviews, the number of 5-star reviews, and the 
# percentage of 5-star reviews for the two types of review (paid vs unpaid)
unpaid_total_reviews = len(vine_program_unpaid)
unpaid_total_reviews
```

```python
unpaid_five_star_reviews = vine_program_unpaid[vine_program_unpaid.star_rating == 5.0]
unpaid_five_star_reviews
```

```python
unpaid_five_star_total = len(unpaid_five_star_reviews)
unpaid_five_star_total
```

```python
percentage_five_star = (unpaid_five_star_total/unpaid_total_reviews) * 100
percentage_five_star
```
